Для улучшения метрики можно преобразовывать входные данные, тюнить параметры логистической регрессии.

```python
# Ваш код подготовки данных здесь. Допишите инициализацию входных фичей и целевую переменную
#Очень долго не мог понять, почему точность предсказания была чуть выше 32%. Далее почитав чат, понял, что нужно удалить
#колонку 'id'. После этого всё стало на свои места
x = df_copy.drop(['odometer/price_std', 'price_category', 'price', 'id'], axis = 1)
y = df_copy.price_category

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
confusion_matrix(y_test, logreg_pred)
```

```python
#Проверяю предсказания на тренировочной выборке
logreg_pr = log_reg.predict(x_train)
print('Точность предсказания по метрике accuracy на тренировочой выборке:', accuracy_score(y_train, logreg_pr))
```

```python
confusion_matrix(y_train, logreg_pr)
```
