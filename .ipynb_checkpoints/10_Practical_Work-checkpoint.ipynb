```python
# Ваш код здесь
power_2_numbers = [2 ** x for x in range(1, 201)]

generated_set = {x**0 if x%5 == 0 else x for x in range (8, 51)}

new_set = set()
power_2_numbers = set(power_2_numbers)

new_set = power_2_numbers.symmetric_difference(generated_set)

```

```python
set_new = set()
for elem in new_set:
    if elem < 34 or elem > 120:
        set_new.add(elem)
new_set = new_set.difference(set_new)
print(new_set)        
```

*Задача 2. Дежурство*

Дан словарь `shedule`, в котором хранится информация о дежурствах. Ключом является id сотрудника, значением — список из дней, в которые данный сотрудник может выйти на дежурство. Обозначения имён всех дней лежат в списке `unique_days`. 

Найдите:

1.   Дни, когда никто не может дежурить (используйте множества для решения этого пункта).
2.   Для каждого дня недели найдите количество сотрудников, которые могут выйти на дежурство в соответствующий день.

Используйте функции и декомпозицию.

```python
shedule = {
    '1001': ['пн', 'вс'],
    '1231': ['пн', 'вт'],
    '1232': ['ср', 'чт', 'пт'],
    '1280': ['ср', 'пт', 'вс'],
    '1282': ['чт'],
    '1290': ['пт', 'вс'],
    '1303': ['вт', 'вс'],
}

unique_days = ['пн', 'вт', 'ср', 'чт', 'пт', 'сб', 'вс']
```

```python
# Ваш код здесь

def check_day(mount_days, unique_days):
    total_days = []
    for i in unique_days:
        day = 0
        for j in mount_days:
            if i == j:
                day += 1
        total_days.append(day)
    
    return total_days


def days_off_duty(mount_days, unique_days):
    all_duty_days = set()
    for days in mount_days:
        all_duty_days.add(days)
    days_with_no_duty = all_duty_days ^ set(unique_days)
    return days_with_no_duty

def results(shedule, unique_days):
    mount_days = []
    values_of_shedule = list(shedule.values())
    mount_days = [x for x in values_of_shedule for x in x]
    total_employee = {unique_day:total_day for unique_day, total_day in zip(unique_days, check_day(mount_days, unique_days))}
    print(f'Дни, когда никто не может дежурить: {days_off_duty(mount_days, unique_days)}'), 
    print(f'Количество сотрудников по дням недели: {total_employee}')
```

```python
results(shedule, unique_days)
```

*Задача 3. Расстояние между двумя точками*

В списке `coords` лежат 2 кортежа, каждый из которых является координатами точек A и B соответственно. Напишите функцию, которая ищет расстояние между этими точками.

Найти расстояние между двумя точками можно по следующим формулам:


Снимок экрана 2021-10-24 в 8.58.08.png

Рекомендации к выполнению:

1.   Используйте функцию `zip`, чтобы итерироваться одновременно по координатам точки A и точки B.
2.   Квадратный корень числа можно вычислить, возведя его в степень 0.5.



```python
coords1 = [(1, 10), (1, 11)]
coords2 = [(0, 20), (20, 20)]
coords3 = [(100, 91, 87), (80, 35, 70)]
coords4 = [(0, 100, 20), (0, 0, 80)]
```

```python
# Ваш код здесь
def distance(a, b):
    dist = []
    for a1, b1 in zip(a, b):
        dist1 = (b1 - a1) ** 2
        dist.append(dist1)
    return dist

def lenght(coords):
    a, b = coords
    total_lenght = (sum(distance(a, b))) ** 0.5
    return total_lenght
```

```python
print(lenght(coords1))
print(lenght(coords2))
print(lenght(coords3))
print(lenght(coords4))
```

*Задача 4. Учёт товаров*

Дано два списка: `ids`, `count`. В первом списке лежат уникальные идентификаторы товаров, во втором списке — количество соответствующих товаров на складе. 

Из данных в списках создайте словарь `items_count`, в котором будет храниться количество товара на складе. Ключ — ID товара, значение — количество. Решение оформите в одну строчку.

ID товара — это строка вида: `число1:число2`. Число1 обозначает категорию товара, а число2 — его номер. Посчитайте следующие статистики на этих данных:
1. Количество уникальных категорий (решение оформите в одну строчку).
2. Среднее количество **единиц товара** на складе в категории (округлите до 1 цифры после запятой).
3. Среднее количество **уникальных товаров** в категории.
4. ID товаров с максимальным количеством единиц на складе.


```python
ids = ['1111:271', '1111:190', '1231:106', '1211:104', '1111:201', '1231:120', '1001:205', '1001:223', '1001:127', '1001:236', \
      '1111:229', '1231:286', '1231:195', '1001:279', '1001:124', '1111:292', '1505:219', '1231:259', '1231:253', '1001:220', '1001:202',\
      '1231:103', '1211:249', '1212:275']
count = [111, 3, 13, 111, 9, 5, 17, 10, 13, 3, 16, 4, 16, 11, 18, 12, 14, 4, 3, 2, 14, 14, 10, 10]
```

```python
# Ваш код здесь/ Словарь items_count, в котором улючи ID товаров, а значения - количество их на складе
items_count = {ids:count for ids, count in zip(ids, count)}
items_count
```

```python
# Количество уникальных категорий
unique_categories = len({id1.split(':')[0].replace(' ', '') for id1 in ids})
unique_categories
```

```python
#Среднее количество единиц товара на складе в категории
print(round(sum(count) / len({k.split(":")[1] for k in items_count.keys()}), 1))
```

```python
#Среднее количество уникальных товаров в категории
print(round(sum(count) / unique_categories, 1))
```

```python
#ID товаров с максимальным количеством единиц на складе
max_ids = [i for i, c in zip(ids, count) if c == max(count)]
max_ids
```
